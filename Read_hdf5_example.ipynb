Reading Vgg16 weights from fastai platform

```python
import h5py

filename = '/home/ariel/DL/tensorflow/tutorials/vgg16.h5'

f = h5py.File(filename, 'r')

keys = f.keys()
N = len(keys)

# list all keys
print keys[:N]

# print key number 0
print keys[0]

# weights - first layer
weights = f[keys[0]].values()[0][:]
print weights
print weights.shape

# biases - first layer
biases = f[keys[0]].values()[1][:]
print biases
```
